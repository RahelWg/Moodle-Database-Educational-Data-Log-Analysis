### Number of quiz submissions by hour of day

```python
%%sql
select count(id),
EXTRACT(HOUR FROM to_timestamp(timecreated)) AS HOUR 
from mdl_logstore_standard_log where action='submitted' AND component='mod_quiz' 
group by EXTRACT(HOUR FROM to_timestamp(timecreated));
```

### Monthly usage time of learners who have confirmed and are not deleted

```python
%%sql
SELECT id,
EXTRACT(MONTH FROM to_timestamp(SUM(lastaccess)-SUM(firstaccess))) AS MONTH,
SUM(lastaccess)-SUM(firstaccess) AS usage_time
FROM mdl_user
WHERE confirmed=1 and deleted=0 
GROUP BY id;
```

### Count of log events per user for the actions

```python
%%sql
SELECT id, count(*)
FROM mdl_logstore_standard_log
WHERE action = 'loggedin'
GROUP BY id
```

```python
%%sql
SELECT id, count(*)
FROM mdl_logstore_standard_log
WHERE action = 'viewed'
GROUP BY id
```

### Overall grade of learners

```python
%%sql
SELECT userid,
ROUND(SUM(finalgrade)/COUNT(*)) AS overall_grade
FROM mdl_grade_grades
WHERE finalgrade IS NOT NULL
GROUP BY userid
```
